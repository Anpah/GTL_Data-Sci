### 1. Plot Activity
Create a plot that displays the data of 4 countries gas prices over the years:<br>

<br>Modify the plot by:

Save each visual in this format: yourname_visualtype.png

```python
gas=pd.read_csv('gas_prices for assignment')
```

```python
gas.head()
```

```python
plt.plot(gas['Year'],gas['Canada'])
plt.plot(gas['Year'],gas['UK'])
plt.plot(gas['Year'],gas['USA'])
plt.plot(gas['Year'],gas['Japan'])
plt.xlabel('Year')
plt.ylabel('Gas Price')
plt.title('Gas Prices of countries Over the Years')

plt.figure(figsize=(10, 6))
for country in gas.columns[1:]:
    plt.plot(gas['Year'], gas[country], label=country)

plt.xlabel('Year')
plt.ylabel('Gas Price')
plt.title('Gas Prices of countries Over the Years')
plt.legend()
plt.savefig('Musfira_gas_prices_plot.png')
plt.show()
```

Improve your plot by:

```python
plt.plot(gas['Year'],gas['Canada'],label='Canada',marker='.',color='r')
plt.plot(gas['Year'],gas['UK'],label='UK',marker='.',color='g')
plt.plot(gas['Year'],gas['USA'],label='USA',marker='.',color='y')
plt.plot(gas['Year'],gas['Japan'],label='Japan',marker='.',color='b')

plt.title('Gas Prices of Countries over the Years')
plt.xlabel('Year')
plt.ylabel('Price per Gallon')
plt.legend()
plt.savefig('Musfira_lineplot.png')

plt.show()
```

### 2. Pie plot activity
<br> Create a pie plot for the data.
- set colors
- explode one section
- indicate their percentages
- legend(give it a title eg Education level)

```python
Workers=[40,35,100,70]
Education=['Primary','JHS','SHS','Tertiary']
```

```python
my_explode = [0,0,0,0.2]
my_colors = ['blue', 'orange', 'green', 'red']

plt.figure(figsize=(6, 6))

plt.pie(Workers, labels=Education,explode=my_explode, autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red'])

plt.title('Education Level of Workers')
plt.legend(title='Education level')
plt.savefig('Musfira_pie_plot.png')
plt.show()
```

### 3. Scatter plot activity
Create a scatter plot with matplotlib that displays xt paired with yt and zt:

```python
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt
```

Change the shape of the points to
1. square for yt
2. triangle for zt
<br> Also:
- Add a label to indicate which points are yt and zt.
- add title
- xlabel and ylabel

```python
plt.figure(figsize=(8, 6))
plt.scatter(xt, yt, label='yt', marker='s')
plt.scatter(xt, zt, label='zt', marker='^')
plt.xlabel('xt')
plt.ylabel('Value')
plt.title('Scatter Plot of xt, yt, and zt')
plt.legend()
plt.savefig('Musfira_scatter_plot.png')
plt.show()
```

### 5. Histogram Activity
Create a histogram of the data below
<br> modify it to your liking

```python
from random import randint
marks=[randint(20,90) for i in range(50)]
bins=[40,50,60,70,80,90,100]
```

```python
plt.figure(figsize=(8, 6))
plt.hist(marks, bins=bins, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')
plt.savefig('Musfira_histogram.png')
plt.show()
```
